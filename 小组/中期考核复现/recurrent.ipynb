```python
import numpy as np
import matplotlib.pyplot as plt

a=np.zeros((15,1000))
#初始化
for i in range(1000):
    a[0][i]=0.01*(i+1)

#opinion的更新
for t in range(1,15):
    b = a
    for i in range(1000):
        nums = 0
        for j in range(1000):
            if(b[t-1][j]-a[t-1][i]<1 and b[t-1][j]-a[t-1][i]>-1):
                a[t][i]+=b[t-1][j]
                nums+=1
        a[t][i]/=nums
a
```

```python
x_axis=np.linspace(0,15,15)
plt.plot(x_axis,a)
plt.title("Fig.1")
plt.xlabel("t")
plt.ylabel("X")
plt.show()
```
